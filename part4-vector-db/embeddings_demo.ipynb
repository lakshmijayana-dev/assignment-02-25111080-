```python
import pandas as pd
df = pd.read_csv("customers.csv")
print(df.head())

documents =df["customer_name"].astype(str).tolist()

sentences = [
    "Cricket is a popular sport in India",
    "Virat Kohli is a great batsman",
    "Bowling requires skill and practice",
    
    "Cooking requires fresh ingredients",
    "I love making biryani at home",
    "Frying food needs proper oil temperature",
    
    "Cybersecurity protects data from hackers",
    "Strong passwords improve security",
    "Phishing attacks are very dangerous",
    
    "Hackers try to steal information online"
]

from sentence_transformers import SentenceTransformer

model = SentenceTransformer('all-MiniLM-L6-v2')
embeddings = model.encode(sentences)


from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(embeddings)
print(similarity)

import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(similarity, annot=True)
plt.title("Similarity Matrix")
plt.show()

query = "The bowler took three wickets in one over"

query_embedding = model.encode([query])
scores = cosine_similarity(query_embedding, embeddings)[0]

top_indices = scores.argsort()[-2:][::-1]

for i in top_indices:
    print(sentences[i], "->", scores[i])


```
